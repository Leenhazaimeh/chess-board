```python
import numpy as np
import matplotlib.pyplot as plt
import math

class ChessBoard:

    def __init__(self):
        self.grid = np.zeros((8,8,3))
        for row in range(8):
            for column in range(8):
                if (row%2 == 0 and column%2 == 0) or (row%2 == 1 and column%2 == 1):
                    self.grid[row,column] = [1,1,1]

 
    
    def add_red(self, row, column):
        self.grid[row, column] =[1,.2,0]

    def add_blue(self, row, column):
        self.grid[row, column] =[0,1,1]

    def render(self):
        plt.imshow(self.grid)
        
 
    def is_under_attack(self, red, blue):
        
        x_dif = abs(red[0] - blue[0])
        y_dif = abs(red[1] - blue[1])
        if (x_dif == 0) or  (y_dif == 0) or (x_dif == y_dif):
            return True
            
        else:
            return False

checkerboared =   ChessBoard() 
checkerboared.add_red(4,2)
checkerboared.add_blue(4,6)
checkerboared.render() 
assert checkerboared.is_under_attack([4,2],[4,6])==True
```

```python
checkerboared =   ChessBoard() 
checkerboared.add_red(4,2)
checkerboared.add_blue(4,3)
checkerboared.render() 
assert checkerboared.is_under_attack([4,2],[4,3])==True
```
